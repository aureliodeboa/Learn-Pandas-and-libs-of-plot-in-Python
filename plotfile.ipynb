Import Libs Matplot e Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Lendo arquivo

```python
filepath ='Iris_Data.csv'

data= pd.read_csv(filepath)
```

```python
plt.plot(data.sepal_length, data.sepal_width, label="sepal", marker='o')
```

Grafico sem as linhas

```python
plt.scatter(data.sepal_length, data.sepal_width, label="sepal", marker='o')
```

Juntado dois tipos de dados no mesmo grafico

```python
plt.scatter(data.sepal_length, data.sepal_width, label="sepal", marker='o')
plt.scatter(data.petal_length,data.petal_width, label="pepal", marker='o',color='red')
```

Histogramas com MatplotLib

```python
plt.hist(data.sepal_length, bins=25)
```

Usando o seabor


```python
import seaborn as sns
```

Grafico de Disperção

```python
sns.jointplot(x='sepal_length',y='sepal_width', data = data, size =4)
```

```python
sns.pairplot(data,hue ='species',size=3)
```
